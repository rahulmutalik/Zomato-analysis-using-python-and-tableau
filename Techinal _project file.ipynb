#### SO there are nan and '-' blank values in the rate column lets replace them 

```python
def handlenew(value):
    if (value=='NEW' or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
df['rate']=df['rate'].apply(handlenew)
```

```python
df['rate'].unique()
```

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
df['rate'].isnull().sum()
```

##### no null values in the rate column.

### Assigning the proper data type to the each of the columns 

```python
df.dtypes
```

```python
df.rename(columns={'approx_cost(for two people)':'2ppl_cost','listed_in(type)':'Type'}, inplace=True)
```

```python
df.head()
```

```python
def handlecomma(value):
    value=str(value)
    if ','in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)
    
df['2ppl_cost']=df['2ppl_cost'].apply(handlecomma)
df['2ppl_cost'].unique()
```

```python
rest_types=df['rest_type'].value_counts(ascending=False)
rest_types
```

```python
rest_types_500=rest_types[rest_types<500]
```

```python
def handle_values(value):
    if (value in rest_types_500):
        return 'others'
    else:
        return value 
df['rest_type']=df['rest_type'].apply(handle_values)
df['rest_type'].value_counts()
```

```python
df=df.drop(['listed_in(city)'],axis=1)
```

```python
df['location'].value_counts()
```

```python
location=df['location'].value_counts()
location_value = location[location<500]

def handle_location(value):
    if(value in location_value):
        return 'others'
    else:
        return value 
df['location']=df['location'].apply(handle_location)
```

```python
df['location'].value_counts()
```

```python
cuisines=df['cuisines'].value_counts()
```

```python
cuisines_less_100=cuisines[cuisines<100]
```

```python
def handle_cusinies(value):
    if (value in cuisines_less_100):
        return 'others'
    else:
        return value 
df['cuisines']=df['cuisines'].apply(handle_cusinies)
```

```python
df['cuisines'].value_counts()
```
